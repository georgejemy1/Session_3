```python
# Define the add function
def add(x, y):
    return x + y

# Define the subtract function
def subtract(x, y):
    return x - y

# Define the division function
def division(x, y):
    # Check if y is zero to avoid division by zero error
    if y == 0:
        print("Cannot divide by zero.")
        return None
    else:
        return x / y

# Define the multiply function
def multiply(x, y):
    return x * y

# Define the even and odd function
def even_and_odd(x):
    # Check if x is an integer
    if isinstance(x, int):
        # Check if x is even or odd using modulo operator
        if x % 2 == 0:
            print(f"{x} is even.")
        else:
            print(f"{x} is odd.")
    else:
        print("Please enter an integer.")

# Define the factorial function
def factorial(x):
    # Check if x is a non-negative integer
    if isinstance(x, int) and x >= 0:
        # Initialize the result variable to 1
        result = 1
        # Use a for loop to calculate the factorial
        for i in range(1, x + 1):
            result = result * i
        return result
    else:
        print("Please enter a non-negative integer.")

# Define the fibonacci function
def fibonacci(x):
    # Check if x is a positive integer
    if isinstance(x, int) and x > 0:
        # Initialize the first two terms of the sequence to 1
        a = 1
        b = 1
        # Use a list to store the sequence
        sequence = [a, b]
        # Use a while loop to generate the sequence until the length is equal to x
        while len(sequence) < x:
            # Calculate the next term by adding the previous two terms
            c = a + b
            # Append the next term to the sequence list
            sequence.append(c)
            # Update the previous two terms
            a = b
            b = c
        return sequence
    else:
        print("Please enter a positive integer.")
```
